```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df= pd.read_excel('Burn-in room temp analysis.xlsx')
df.head()
```

```python
g=sb.PairGrid(data=df, vars =['溫度','佔用oven台次','廠區'])
g=g.map_diag(plt.hist, bins=5);
g.map_offdiag(plt.scatter)
```

```python
sb.pointplot(data=df, x='溫度', y='佔用oven台次', hue='廠區', linestyles='')
plt.show()
```

```python
g=sb.FacetGrid(data=df, hue='廠區', size=8)
g.map(plt.scatter, '溫度', '佔用oven台次')
g.add_legend()
```
