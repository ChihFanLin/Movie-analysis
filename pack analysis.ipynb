```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df= pd.read_excel('包裝出貨.xlsx')
df.head()
```

```python
df.describe()
```

```python
# correlation plot
plt.figure(figsize=[12,8])
sb.heatmap(df.corr(),annot=True, fmt='.2f',cmap='vlag_r',center=0)
plt.xticks(rotation=90)
plt.show()
```
